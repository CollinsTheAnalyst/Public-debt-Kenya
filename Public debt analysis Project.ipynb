```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.interpolate import make_interp_spline

# Ensure 'Date' is in datetime format (if it's not already)
debt['Date'] = pd.to_datetime(debt['Date'])

# Sort the data by 'Date' to ensure it's in ascending order
debt = debt.sort_values('Date').drop_duplicates(subset='Date')

# Set the figure size
plt.figure(figsize=(12, 6))

# Convert 'Date' to numerical values for interpolation
debt['Date_numeric'] = debt['Date'].map(pd.Timestamp.toordinal)

# Create smoother lines using interpolation for Domestic Debt
x_smooth = np.linspace(debt['Date_numeric'].min(), debt['Date_numeric'].max(), 500)
spl_domestic = make_interp_spline(debt['Date_numeric'], debt['Domestic Debt'], k=3)
y_smooth_domestic = spl_domestic(x_smooth)

# Create smoother lines using interpolation for External Debt
spl_external = make_interp_spline(debt['Date_numeric'], debt['External Debt'], k=3)
y_smooth_external = spl_external(x_smooth)

# Convert x_smooth back to datetime using ordinals
x_smooth_dates = [pd.Timestamp.fromordinal(int(date)) for date in x_smooth]

# Plot the smooth lines
plt.plot(x_smooth_dates, y_smooth_domestic, label='Domestic Debt', color='blue')
plt.plot(x_smooth_dates, y_smooth_external, label='External Debt', color='orange')

# Add titles and labels
plt.title('Trends in Domestic and External Debt Over Time', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Debt Amount (in currency)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by year and calculate total domestic and external debt
debt['Year'] = debt['Date'].dt.year
annual_debt = debt.groupby('Year').agg(
    Total_Domestic_Debt=('Domestic Debt', 'sum'),
    Total_External_Debt=('External Debt', 'sum')
).reset_index()

# Calculate the Debt Ratio for each year
annual_debt['Debt Ratio'] = annual_debt['Total_Domestic_Debt'] / annual_debt['Total_External_Debt']

# Set the figure size
plt.figure(figsize=(12, 6))

# Plot Annual Debt Ratio
plt.plot(annual_debt['Year'], annual_debt['Debt Ratio'], label='Domestic to External Debt Ratio', marker='o', color='purple')

# Set x-axis ticks with a difference of 2 years
plt.xticks(ticks=annual_debt['Year'][::2])  # Show every 2nd year

# Add titles and labels
plt.title('Annual Domestic to External Debt Ratio', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Debt Ratio')
plt.axhline(y=1, color='r', linestyle='--', label='Ratio = 1 (Equal Debt)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()

# Print the annual ratio values
print(annual_debt[['Year', 'Debt Ratio']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create a figure
plt.figure(figsize=(12, 5.5))

# Define colors based on the value of 'Annual GDP growth (%)'
def get_color(value):
    if value < 0:
        return 'red'         # Negative values
    elif 0 <= value <= 3.5:
        return 'yellow'      # Values between 0 and 3.5
    else:
        return 'green'     # Values above 3.5

# Apply the color function to create an array of colors
colors = [get_color(value) for value in gdp['Annual GDP growth (%)']]

# Create the bar plot with the specified colors
sns.barplot(x='Year', y='Annual GDP growth (%)', data=gdp, palette=colors)

# Add titles and labels
plt.title('Annual GDP Growth (%) Trend', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('GDP Growth (%)')

# Add data labels on top of the bars
for index, value in enumerate(gdp['Annual GDP growth (%)']):
    plt.text(index, value + 0.2 if value > 0 else value - 0.2, f'{value:.1f}%', 
             ha='center', va='bottom' if value > 0 else 'top', color='black')

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV files
debt = pd.read_csv(r"C:\Users\admin\Downloads\Public Debt.csv")
percapita = pd.read_csv(r"C:\Users\admin\Downloads\per-capita.csv")

# Make sure the 'Year' column in the debt DataFrame is in the correct format
debt['Year'] = pd.to_datetime(debt['Year'].astype(str) + ' ' + debt['Month']).dt.year

# Display the unique values in the 'Year' columns for both DataFrames
print("\nUnique Years in Debt DataFrame:", debt['Year'].unique())
print("Unique Years in Per Capita DataFrame:", percapita['date'].unique())

# Rename 'date' to 'Year' in percapita for consistency
percapita.rename(columns={'date': 'Year'}, inplace=True)

# Convert 'Year' in percapita to integer
percapita['Year'] = percapita['Year'].astype(int)

# Filter out the rows for June in the debt DataFrame and calculate % growth in total debt
debt_june = debt[debt['Month'] == 'June'].copy()
debt_june['% Growth in Total Debt'] = debt_june['Total'].pct_change() * 100

# Merge the two DataFrames on 'Year'
merged_df = pd.merge(percapita, debt_june[['Year', '% Growth in Total Debt']], on='Year', how='left')

# Plotting
plt.figure(figsize=(12, 6))

# Plotting % Growth in GDP Per Capita
plt.plot(merged_df['Year'], merged_df['Annual per capita Growth Rate(%)(gdp)'], marker='o', label='GDP Per Capita Growth Rate (%)', color='green')

# Plotting % Growth in Debt Per Capita
plt.plot(merged_df['Year'], merged_df['Annual per capita Growth Rate(%)(debt)'], marker='o', label='Debt Per Capita Growth Rate (%)', color='orange')

# Creating a secondary y-axis for % Growth in Total Debt as a bar chart
ax2 = plt.gca().twinx()  # Twin Axes sharing the x-axis
ax2.bar(merged_df['Year'], merged_df['% Growth in Total Debt'], color='blue', alpha=1, label='% Growth in Total Debt', width=0.4)

# Adding titles and labels
plt.title('Growth Rates: GDP Per Capita, Debt Per Capita, and Total Debt (%)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Growth Rate (%)', fontsize=14)

# Setting label for the second y-axis
ax2.set_ylabel('% Growth in Total Debt', fontsize=14, color='red')

# Adding legends
plt.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()  # To adjust the layout for better visibility
plt.show()
```
